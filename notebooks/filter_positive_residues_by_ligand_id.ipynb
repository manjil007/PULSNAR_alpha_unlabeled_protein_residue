```python
import pathlib
import pandas as pd
```

```python
# --- 1. Load the full dataframe ----------------------------------------
df = pd.read_csv('../data/processed/BioLiP_positives_residue_level_with_duplicates_1.csv')

# --- 2. Output folder ---------------------------------------------------
out_dir = pathlib.Path('../data/processed/ligand_splits')
out_dir.mkdir(parents=True, exist_ok=True)

# --- 3. Threshold -------------------------------------------------------
MIN_ROWS = 5_000          # change this if you like

# --- 4. One file per ligand_id (only if  MIN_ROWS) --------------------
for ligand_id, subset in df.groupby('ligand_id'):
    if len(subset) >= MIN_ROWS:
        out_path = out_dir / f'positives_{ligand_id}.csv'
        subset.to_csv(out_path, index=False)
        print(f'Wrote {len(subset):,} rows  {out_path}')
    else:
        print(f'Skipped {ligand_id!r}: only {len(subset):,} rows')
```
